```python
import pandas as pd
obs = pd.read_csv('observations.csv')
spec = pd.read_csv('species_info.csv')
```

```python
obs.head()
```

```python
spec.head()
```

```python
obs_sci = obs['scientific_name'].tolist()
spec_sci = spec['scientific_name'].tolist()
spec_comm = spec['common_names'].tolist()
```

```python
common_obs = []
for i in obs_sci:
    if i in spec_sci:
        index = spec_sci.index(i)
        common_obs.append(spec_comm[index])
    else:
        common_obs.append("NaN")
```

```python
print(common_obs[0:5])
```

```python
obs['common_name'] = pd.Series(common_obs)
```

```python
obs.head()
```

```python
obs.park_name.unique()
```

```python
gsmoky = []
gsmoky2 = []
yosemite = []
yosemite2 = []
bryce = []
bryce2 = []
yellowstone = []
yellowstone2 = []

for i in range(len(obs.observations.tolist())):
    if obs.park_name.iloc[i]=="Great Smoky Mountains National Park":
        gsmoky.append(obs.common_name.iloc[i])
        gsmoky2.append(obs.observations.iloc[i])
    elif obs.park_name.iloc[i]=="Yosemite National Park":
        yosemite.append(obs.common_name.iloc[i])
        yosemite2.append(obs.observations.iloc[i])
    elif obs.park_name.iloc[i]=="Bryce National Park":
        bryce.append(obs.common_name.iloc[i])
        bryce2.append(obs.observations.iloc[i])
    elif obs.park_name.iloc[i]=="Yellowstone National Park":
        yellowstone.append(obs.common_name.iloc[i])
        yellowstone2.append(obs.observations.iloc[i])
```

```python
import matplotlib.pyplot as plt
x = gsmoky[0:20]
y = gsmoky2[0:20]
plt.figure(figsize = (100, 7))
plt.title("Wildlife Observed in Great Smoky Mountains National Park")
plt.bar(x, y)
plt.show()

print("Double Click to expand chart")
```

```python
import matplotlib.pyplot as plt
x = yosemite[0:20]
y = yosemite2[0:20]
plt.figure(figsize = (100, 7))
plt.title("Wildlife Observed in Yosemite National Park")
plt.bar(x, y)
plt.show()

print("Double Click to expand chart")
```

```python
import matplotlib.pyplot as plt
x = bryce[0:20]
y = bryce2[0:20]
plt.figure(figsize = (100, 7))
plt.title("Wildlife Observed in Bryce National Park")
plt.bar(x, y)
plt.show()

print("Double Click to expand chart")
```

```python
import matplotlib.pyplot as plt
x = yellowstone[0:20]
y = yellowstone2[0:20]
plt.figure(figsize = (100, 7))
plt.title("Wildlife Observed in Yellowstone National Park")
plt.bar(x, y)
plt.show()

print("Double Click to expand chart")
```

```python
gsmoky_total = sum(gsmoky2)
yosemite_total = sum(yosemite2)
bryce_total = sum(bryce2)
yellowstone_total = sum(yellowstone2)

totals_y = [gsmoky_total, yosemite_total, bryce_total, yellowstone_total]
x = ["Great Smoky Mountains", "Yosemite", "Bryce National Park", "Yellowstone"]

plt.figure(figsize = (10, 5))
plt.bar(x, totals_y)
plt.title("Total number of wildlife observed in each national park")
plt.show()
```

# Analysis

## Although the distribution of wildlife observations seemd relatively similar for each national park, when comparing how many total observations each park gathered, there was a significant difference. Yellowstone national park had much more total observations of different wildlife than all the other parks.
